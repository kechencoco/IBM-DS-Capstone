# The Battle of Neighborhoods - Project Report

## 1. Introduction

When people are looking for proper neighborhood, they will consider different factors like price, education, and living of convenience, they will consider whether there are supermarkets, grocery stores, coffee shops, hospitals or cinema nearby.

The reason we choose Scarborough is because it is one of the most diverse and multicultural areas in the Greater Toronto Area since it is becoming a popular destination for new immigrants in Canada to live in.


## 2. Business Problem

This project is to create an analysis of features aiming to help those people who are willing to rent or buy apartment in Scarborough, Toronto to better make their decisions. It will help them to explore facilities around the neighborhood there. The feature in this analysis will include housing price, school ratings, crime rates, weather, road connectivity, management for emergency, water resource and recreational facilities.

It will help people to make better decision and get fully to know the area and neighborhood before renting or buying an apartment or a house

## 3. Foursquare API

This project will use Four-square API as main data source since the places API offer the ability to perform location search, location sharing and details about a business.

The data retrieved from Foursquare contained information of venues within a specified distance of the longitude and latitude of the postcodes. The information obtained per venue as follows:

1. Neighborhood
2. Neighborhood Latitude
3. Neighborhood Longitude
4. Venue
5. Name of the venue e.g. the name of a store or restaurant
6. Venue Latitude
7. Venue Longitude
8. Venue Category
